```python
url = '/content/dataset_CD_UC/ulcerative-colitis/0954d7e1-cbbc-4b1e-b0db-e5db9a3579fd.jpg'
    
img = imread(url)
```

```python
B = crop(img, ((0, 173), (25, 0), (0,0)), copy=False)

#imsave('/content/dataset_CD_UC/ulcerative-colitis/0954d7e1-cbbc-4b1e-b0db-e5db9a3579fd.jpg', B)
```

```python
C = crop(img, ((0, 25), (210, 0), (0,0)), copy=False)
print(img.shape, B.shape, C.shape)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,15))
ax1.imshow(img)
ax2.imshow(B)
ax3.imshow(C)
ax1.set_title('Original image')
ax2.set_title('Horizontal Crop image')
ax3.set_title('Vertical Crop image')
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')
```

<h1>Blur</h1>

```python
url = '/content/dataset_CD_UC/ulcerative-colitis/d5570649-681f-4b6d-816b-6a3140f64b68.jpg'
img = imread(url)
```

```python
blur_img = img.copy()
blur_img[350:,0:210] = np.dot(2/16,blur_img[350:,0:210])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,15))
ax1.imshow(img)
ax2.imshow(blur_img)
ax1.set_title('Original image')
ax2.set_title('Blur image')
ax1.axis('off')
ax2.axis('off')
```
